# **6. Data Cleansing-Standarisasi Kolom brand**

```python
df['brand'] = df['brand'].str.strip().str.title() #strip untuk membersihkan spasi + title untuk standarisasi huruf (mengubah menjadi huruf kapital diawal kata) sekaligus mengembalikan data yang sudah distandarisasi kedalam kolom brand

print(df['brand'].unique()[:20]) #menampilkan 20 data unik (tidak duplikat)
```

# **7. Data Cleansing-Standarisasi Kolom perfume**

```python
df['perfume'] = df['perfume'].str.strip().str.title() #strip untuk membersihkan spasi + title untuk standarisasi huruf (mengubah menjadi huruf kapital diawal kata) sekaligus mengembalikan data yang sudah distandarisasi kedalam kolom perfume

print(df['perfume'].unique()[:10]) #menampilkan 10 data unik (tidak duplikat)
```

# **8. Data Cleansing-Standarisasi Kolom type**

```python
df['type'] = df['type'].str.strip().str.title() #strip untuk membersihkan spasi + title untuk standarisasi huruf (mengubah menjadi huruf kapital diawal kata) sekaligus mengembalikan data yang sudah distandarisasi kedalam kolom type

print(df['type'].unique()[:10]) #menampilkan 10 data unik (tidak duplikat)
```

# **9. Data Cleansing-Standarisasi Kolom category**

```python
df['category'] = df['category'].str.strip().str.title() #strip untuk membersihkan spasi + title untuk standarisasi huruf (mengubah menjadi huruf kapital diawal kata) sekaligus mengembalikan data yang sudah distandarisasi kedalam kolom category

print(df['category'].unique()) #menampilkan data unik (tidak duplikat)
```

# **10. Data Cleansing-Standarisasi Kolom target_audience**

```python
df['target_audience'] = df['target_audience'].str.strip().str.title() #strip untuk membersihkan spasi + title untuk standarisasi huruf (mengubah menjadi huruf kapital diawal kata) sekaligus mengembalikan data yang sudah distandarisasi kedalam kolom target_audience

print(df['target_audience'].unique()) #menampilkan data unik (tidak duplikat)
```

# **11. Data Cleansing-Standarisasi Kolom longevity**

```python
def clean_longevity(value):
    if not isinstance(value, str): #mengecheck nilai string
        return value

    value = re.sub(r":contentreference\[oaicite:\d+\]\{index=\d+\}", "", value, flags=re.IGNORECASE) #menghapus yang tidak sesuai, d+ artinya angkanya =1 atau lebih, 'ignorecase' untuk mengabaikan kata tersebut memiliki huruf besar atau kecil akan tetap dihapus asalkan kata yang dihapus sama

    value = re.sub(r"\s+", " ", value) #merapikan spasi
    return value.strip()

df['longevity'] = df['longevity'].apply(clean_longevity) #mengembalikan data yang sudah distandarisasi kedalam kolom longecity

print(df['longevity'].unique()) #menampilkan data unik (tidak duplikat)
```

# **12. Data Cleansing-Deduplikasi Data**

```python
duplicate_rows = df[df.duplicated(['brand', 'perfume', 'type', 'category', 'target_audience', 'longevity'], keep='first')]
print("Jumlah data yang duplikat:", len(duplicate_rows))
duplicate_rows #menampilkan data yang duplikat
```

# **13. Data Cleansing-Data Setelah Cleansing Duplikasi**

```python
df_cleaned = df.drop_duplicates(['brand', 'perfume', 'type', 'category', 'target_audience', 'longevity'], keep='first')
print("\nJumlah data setelah cleansing:", len(df_cleaned))
df_cleaned #menampilkan data bersih
```

# **14. Data Enrichment**

```python
df_cleaned['product_id'] = ['PRF' + str(i+1).zfill(4) for i in range(len(df_cleaned))]

df_cleaned
```
